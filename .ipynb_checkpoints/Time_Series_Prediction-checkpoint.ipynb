# Predicting Random Wave data using Time Series Analysis

## Import the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, LSTM, Bidirectional
```

## Create the dataset
We will be using `sine`, `cosine` and `random` functions to get a random data

```python
X_train = np.arange(0, 100, 0.1)
y_train = np.sin(X_train)
y_train += np.cos(-X_train)
y_train += np.random.random(1000)
```

## Data Analysis

```python
plt.figure(figsize=(10, 4))
plt.grid()
plt.plot(X_train, y_train)
plt.show()
```

```python
X_train.shape, y_train.shape
```

```python
min(y_train), max(y_train)
```

## Data Preperation
* Split the data into `training` and `validation` sets with the ratio 80:20
* We will be using `TimeseriesGenerator` module from `tensorflow` for getting a timeseries format dataset.
* We will be using sliding window technique with `window_size = 20` and `slide_length = 1`. i.e., given `x1, x2, x3, ..., x20` predict the `x21` value.

```python
X_val = X_train[int(len(X_train)* 0.8):]
X_train = X_train[:int(len(X_train)* 0.8)]

X_test = X_val[int(len(X_val)* 0.5):]
X_val = X_val[:int(len(X_val)* 0.5)]

y_val = y_train[int(len(y_train)* 0.8):]
y_train = y_train[:int(len(y_train)* 0.8)]

y_test = y_val[int(len(y_val)* 0.5):]
y_val = y_val[:int(len(y_val)* 0.5)]
```

```python
len(X_train), len(X_val), len(X_test), len(y_train), len(y_val), len(y_test)
```

```python
plt.figure(figsize=(10, 4))
plt.grid()
plt.plot(X_train, y_train, c='b', label='train')
plt.plot(X_val, y_val, c='r', label='valid')
plt.plot(X_test, y_test, c='g', label='test')
plt.legend()
```

```python
n_features = 1
window_size  = 20
batch_size = 10
sampling_rate = 1
slide_length = 1

train_data = y_train.reshape((len(y_train), n_features))
val_data = y_val.reshape((len(y_val), n_features))
test_data = y_test.reshape((len(y_test), n_features))

train_generator = TimeseriesGenerator(train_data, train_data,
                                      length        = window_size,
                                      sampling_rate = sampling_rate,
                                      stride        = slide_length,
                                      batch_size    = batch_size)

val_generator = TimeseriesGenerator(val_data, val_data,
                                      length        = window_size,
                                      sampling_rate = sampling_rate,
                                      stride        = slide_length,
                                      batch_size    = batch_size)

test_generator = TimeseriesGenerator(test_data, test_data,
                                      length        = window_size,
                                      sampling_rate = sampling_rate,
                                      stride        = slide_length,
                                      batch_size    = batch_size)
```

## Tensorflow Model creation
* We will be creating a simple model with `two` LSTM layers and `one` output layer.
* LSTM layer will have units `8` and `16` for first and second LSTM layers respectively.
* Output layer will gave `1` unit as output. No activation will be used.
* `Adam` optimizer with a learning_rate of `0.001` is used.
* `Mean Squared Error` is used as loss function.

```python
tf.random.set_seed(42)

units  = 8

model = Sequential()

model.add(Input(shape=(window_size, n_features), name='input_layer'))
model.add(LSTM(units, return_sequences=True, name='lstm_layer_1'))
model.add(LSTM(units * 2, return_sequences=True, name='lstm_layer_2'))
model.add(LSTM(units * 4, name='lstm_layer_3'))
model.add(Dense(1, name='prediction_layer'))

model.compile(optimizer='adam', loss='mse')

model.summary()


history = model.fit(train_generator, validation_data=val_generator, epochs=500, verbose=0)
```

## Evaluate Model
* Evaluate the model on Validation and Test Dataset

```python
val_predictions  = model.predict(val_generator)
model.evaluate(val_generator)
```

```python
test_predictions  = model.predict(test_generator)
model.evaluate(test_generator)
```

## Plot the predictions

```python
plt.figure(figsize=(10, 4))
plt.grid()
plt.plot(X_train, y_train, c='b', label='train')
plt.plot(X_val[:window_size], y_val[:window_size], c='r', label='valid')
plt.scatter(X_val[window_size:], val_predictions, c='y', s=3, label='valid_pred')
plt.plot(X_test[:window_size], y_test[:window_size], c='g', label='test')
plt.scatter(X_test[window_size:], test_predictions, c='m', s=3, alpha=0.3, label='test_pred')
plt.legend()
```

```python
model.save('time_series_analizer.keras')
```
